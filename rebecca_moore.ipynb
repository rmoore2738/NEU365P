```python
animals = ['Orca', 'Wolf', 'Gecko']
animals[1] = 'Dog'
animals.extend(['Puma', 'Jaguar'])
animals
```

```python
cats = animals[3:]
cats
```

```python
import copy
animalsRef = animals
animalsCopy = copy.copy(animals)
animalsRef[1] = 'Wolf'
animalsCopy[0] = 'Dolphin'
print(animals)
print(animalsRef)
print(animalsCopy)
```

```python
for i in animals:
    if i in cats:
        print(i)
```

```python
for (i, val) in enumerate(animals):
    if val == 'Gecko':
        animals[i] = 'Salamander'
print(animals)
```
