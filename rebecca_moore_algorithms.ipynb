3. (3 pts) Write a function implementing the selection sort algorithm on an input list.

```python
def selection_sort(items):
    for i in range (len(items) - 1):
        min = items[i]
        minIdx = i
        for j in range (i + 1, len(items)):
          if (items[j] < min):
            min = items[j]
            minIdx = j
        items[minIdx] = items[i]
        items[i] = min
    return items
```
