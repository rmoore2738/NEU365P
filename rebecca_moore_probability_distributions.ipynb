```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import scipy.stats as st

rate = 2.4
dist = st.poisson(rate)
numMutations = np.arange(11)

plt.bar(numMutations, dist.pmf(numMutations))
plt.title('Probabilities of 0-10 mutations', fontsize = 16)
plt.xlabel('# mutations', fontsize = 16)
plt.ylabel('probability', fontsize = 16);

print(f"Deaths expected out of 1000 residents = {(1000 * 0.003338616949228422) / 2}")
print()
```

```python
rate = 0.5
dist = st.expon(loc = 0, scale = 1 / rate)
high = dist.ppf(.95)
intervalsSec = np.linspace(0, high, 501)

pdf = dist.pdf(intervalsSec)

plt.plot(intervalsSec, pdf)
plt.title('Distribution of interspike intervals', fontsize = 16)
plt.xlabel('interspike interval (sec)', fontsize = 16)
plt.ylabel('pdf', fontsize = 16);

dist1 = st.erlang(a = 2, scale = 1 / rate)
print(f'Average spike rate of downstream neuron: {dist.cdf(0.1)}')
print()
```

```python
n = 5
p = 0.30
dist = st.binom(n, p)
dist.pmf(4)

ligand_bound = np.arange(n)
pmf = dist.pmf(ligand_bound)

plt.bar(ligand_bound, pmf)
plt.title('Probability of ligand bound at given concentration', fontsize = 16)
plt.xlabel('# ligand bound', fontsize = 16)
plt.ylabel('Probability', fontsize = 16)
```

```python
mu = 64
sd = 12
dist = st.norm(mu, sd)

scores = np.linspace(20, 120, 301)
pdf = dist.pdf(scores)

plt.plot(scores, pdf)
plt.title('Distribution of scores', fontsize = 16)
plt.xlabel('scores', fontsize = 16)
plt.ylabel('probability of score', fontsize = 16);

print(f"The probability that a mouse is depressed = {dist.sf(80)}")
print()
```
